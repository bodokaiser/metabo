```python
import numpy as np
import scipy as sp
```

```python
import nmrglue as ng

acetone1 = ng.varian.read('HMDB01659-1H')
acetone2 = ng.bruker.read('HMDB01659-13C')
```

```python
from matplotlib import pyplot as plt

fig, axes = plt.subplots(2, sharex=True, figsize=(12, 12))

axes[0].plot(acetone1[1])
axes[1].plot(sp.fft(acetone1[1]))

plt.show()
```

```python
fig, axes = plt.subplots(2, sharex=True, figsize=(12, 12))

axes[0].plot(acetone2[1])
axes[1].plot(sp.fft(acetone2[1]))

plt.show()
```
