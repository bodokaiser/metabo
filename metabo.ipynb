```python
import numpy as np
import scipy as sp
import nmrglue as ng
```

```python
from matplotlib import pyplot as plt

acetone = ng.varian.read('HMDB01659-1H')

fig, axes = plt.subplots(2, sharex=True, figsize=(12, 12))

axes[0].plot(acetone[1])
axes[1].plot(np.abs(sp.fft(acetone[1])))

plt.show()
```
