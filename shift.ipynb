```python
import numpy as np
import scipy as sp
import nmrglue as ng

from matplotlib import pyplot as plt
```

```python
dic1, fid1 = ng.varian.read('HMDB00122-1H')
dic2, fid2 = ng.bruker.read('HMDB01875-1H')
```

```python
udic1 = ng.varian.guess_udic(dic1, fid1)
udic2 = ng.bruker.guess_udic(dic2, fid2)

uc1 = ng.fileiobase.uc_from_udic(udic1)
uc2 = ng.fileiobase.uc_from_udic(udic2)
```

```python
plt.plot(uc1.ppm_scale(), np.abs(np.real(sp.fft(fid1))))
plt.show()
```

```python
plt.plot(uc2.ppm_scale(), np.abs(np.real(sp.fft(fid2))))
plt.show()
```
